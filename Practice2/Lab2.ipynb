```python
n = int(input())

if n <= 1:
    print("No")
else:
    is_prime = True
    for i in range(2, n):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        print("Yes")
    else:
        print("No")
```

```python
n = int(input())
arr = list(map(int, input().split()))
t = 0
result = arr[0]
for i in arr:
    count = 0
    for j in arr:
        if i == j:
            count += 1
    if (count > t) or (count == t and i < result):
        t = count
        result = i

print(result)
```

```python
n = int(input())
numbers = [input().strip() for _ in range(n)]
result = 0
checked = []

for num in numbers:
    if num not in checked:
        if numbers.count(num) == 3:
            result += 1
        checked.append(num)

print(result)
```

```python
n = int(input())
arr = [input().strip() for _ in range(n)]
first_index = {}
for i, s in enumerate(arr):
    if s not in first_index:
        first_index[s] = i + 1 
for s in sorted(first_index):
    print(s, first_index[s])
```

```python
n = int(input())
episodes = {}
for _ in range(n):
    line = input().split()
    name = " ".join(line[:-1])
    count = int(line[-1]) 
    if name in episodes:
        episodes[name] += count
    else:
        episodes[name] = count
for dorama in sorted(episodes):
    print(dorama, episodes[dorama])
```

```python
import sys
readline = sys.stdin.readline
write = sys.stdout.write
n = int(readline())
db = {}
for _ in range(n):
    parts = readline().rstrip('\n').split(' ',2)
    if parts[0] == 'set':
        key = parts[1]
        value = parts[2]
        db[key] = value
    else:
        key = parts[1]
        if key in db:
            write (db[key] + "\n")
        else:
            write(f"KE: no key {key} found in the document\n")
```
